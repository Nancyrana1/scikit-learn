```python
condition=[
    train['PRICE']<10,
    (train['PRICE']>=10) & (train['PRICE']<20),
    (train['PRICE']>=20) & (train['PRICE']<30),
    (train['PRICE']>=30) & (train['PRICE']<40),
    train['PRICE']>=40
]
choice=[1,2,3,4,5]

train['category'] = np.select(condition, choice)

category_counts = train['category'].value_counts()

print(category_counts)
```

```python
import numpy as np

train['RM'] = train['RM'].replace(0,np.nan)
train['RIVERSIDE'] = train['RIVERSIDE'].replace('UNKNOWN',np.nan)
train['AGE'] = train['AGE'].replace('-1',np.nan)
```

```python
X=train.drop(['PRICE'],axis=1)
y = train['PRICE']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape)
```

```python
print(X_train.columns)

print(X_val.columns)
```

```python
X_train =X_train.drop(['category'],axis=1)
X_val =X_val.drop(['category'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('crim', MinMaxScaler(), ['CRIM']),
        ('zn', MinMaxScaler(), ['ZN']),
        ('indus', StandardScaler(), ['INDUS']),
        ('polindex', MinMaxScaler(), ['POLINDEX']),
        ('dis', MinMaxScaler(), ['DIS']),
        ('hw', MinMaxScaler(), ['HIGHWAYCOUNT']),
        ('tax', MinMaxScaler(), ['TAX']),
        ('ptr', MinMaxScaler(), ['PTRATIO']),
        ('imm', MinMaxScaler(), ['IMM']),
        ('bpl', MinMaxScaler(), ['BPL']),

        ('rm', Pipeline([
            ('imputer' , SimpleImputer(strategy='median')),
            ('scaler', MinMaxScaler())
        ]),['RM']),
        ('age', Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('scaler', MinMaxScaler())
        ]), ['AGE']),

        ('river', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder())
        ]), ['RIVERSIDE'])

    ]
)
```

```python
x_train_transformed = preprocessor.fit_transform(X_train)
x_test_transformed = preprocessor.transform(X_val)
```

```python
mean_val=x_test_transformed.mean()

print(mean_val)
```
